<h1><span style="color:purple">2.1 Ecuaciones diferenciales autónomas<span></h1>

<h2><span style="color:blue" >2.1.1 Definición<span></h2>

Una EDO autónoma de primer orden presenta la forma:

$$
\begin{equation}
\begin{aligned}
\frac{dy}{dx}= f(y)
\end{aligned}
\label{ODE_auto}
\tag{1}
\end{equation}
$$

Las característica fundamental de la ecuación ([1](#mjx-eqn-ODE_auto)) es que la variable independiente, en este caso $x$, <b>no aparece en el lado derecho de la ecuación</b>.

<h2><span style="color:blue" >2.1.2 Soluciones de equilibrio y análisis cualitativo <span></h2>

<h3><span style="color:green"><b>a. Solución constante </b></span></h3>

Si una solución de la EDO autónoma de primer orden (ecuación ([1](#mjx-eqn-ODE_auto))) tiene la forma $y(x)=c$, se dice que esta es una solución constante.

Ahoya, note que $\frac{dy}{dx}=0$ para la solución constante. Por lo tanto, <b>cualquier solución constante debe satisfacer la forma:</b>
$$
\begin{equation}
\begin{aligned}
f(y)=0
\end{aligned}
\label{ODE_auto_1}
\tag{2}
\end{equation}
$$

es decir, las raíces de la función $f(y)$ constituyen soluciones constantes de la EDO autónoma de primer orden $\frac{dy}{dx}=f(y)$. 

Por ejemplo, considere la EDO autónoma de primer orden:
$$
\begin{equation}
\begin{aligned}
\frac{dy}{dx}= y^{2}-1
\end{aligned}
\label{ODE_auto_3}
\tag{3}
\end{equation}
$$

De la ecuación ([3](#mjx-eqn-ODE_auto_3)) podemos distinguir que $f(y)=y^{2}-1$. 

Ahora, según lo especifica la ecuación ([2](#mjx-eqn-ODE_auto_2)), las raíces de la ecuación $f(y)=0$ deben porporcionar las soluciones constantes. Por tanto, resolviendo la ecuación algebraica $y^{2}-1=0$, se obtienen las soluciones constantes $y_{1}(x)=1$ y $y_{2}(x)=-1$.

<h3><span style="color:green"><b>b. Curvas solución </b></span></h3>

Para analizar las características de las curvas solución de una EDO autónoma de primer orden, considere la siguiente ecuación diferencial:

$$
\begin{equation}
\begin{aligned}
\frac{dy}{dx}= 4y(1-y)
\end{aligned}
\label{ODE_auto_4}
\tag{4}
\end{equation}
$$

De la ecuación ([4](#mjx-eqn-ODE_auto_4)) se reconoce fácilmente que $f(y)=4y(1-y)$. Ahora, para obtener las raíces de esta última expresión, hacemos que $f(y)=0$, es decir, $4y(1-y)=0$. De este última expresión, se deduce que la EDO autónoma de primer orden ([4](#mjx-eqn-ODE_auto_4)) tiene dos soluciones constantes: $y_{1}=0$ y $y_{2}=1$. 

```python

import numpy as np
from  scipy.integrate import odeint
%matplotlib inline
#%matplotlib widget
import matplotlib.pyplot as plt
from matplotlib import cm
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import animation
```

```python
def f(x,c):
    return (c*np.exp(4*x))/(1+c*np.exp(4*x))

x = np.linspace(-4,4,500)
c = np.linspace(0,10,15)

def g(Y,X):
    return 4*Y*(1-Y)
X,Y = np.meshgrid(np.linspace(-1.5,1.5,15),np.linspace(-0.5,1.5,15))
#vectorizing the color map
S_values = np.arange(0.0,5,0.25, dtype=float)
vcolors1 = plt.cm.viridis_r(np.linspace(0,0.5, len(S_values))) 
vcolors2 = plt.cm.viridis_r(np.linspace(0.5,1, len(S_values))) 
# colors      for each trajectory
U,V = 1,g(Y,X)
N = np.sqrt(U**2 + V**2)
U /= N
V /= N
plt.quiver(X, Y, U, V ,color = vcolors1,scale=20,linewidth=0.05)
#plt.hlines(1,-1.5,1)
plt.text(-1.3,1.25,r'$R_{1}$', fontsize=18)
plt.text(-1.3,0.60,r'$R_{2}$', fontsize=18)
plt.text(-1.3,-0.25,r'$R_{3}$', fontsize=18)

for i in range(len(c)):
    y = [f(j,c[i]) for j in x]
    plt.plot(x,y,label = 'c='+ str(c[i]), color=vcolors2[i])
#plt.legend()
plt.xlim(-1.5,1.5)
plt.ylim(-0.5,1.5)
plt.xlabel('x', size=14)
plt.ylabel('y(x)', size=14)
plt.title ('Fig. 1: Campo de direcciones y soluciones de la EDO autónoma 4')
plt.show()


```

La Figura 1 muestra el campo de direcciones y un  grupo de curvas solución obtenidas de la familia de soluciones de la ecuación diferencial ([4](#mjx-eqn-ODE_auto_4)), es decir, de $y(x)=\frac{ce^{4x}}{1+ce^{4x}}$. Esta Figura ha sido dividida en tres subregiones: $1<R_{1}<\infty$, $0<R_{2}<1$, $-\infty<R_{3}<0$. Con esto, es posible obtener las siguientes conclusiones:

<ul>
<li>El conjunto de soluciones provenientes de la familia uniparamétrica $y(x)=\frac{ce^{4x}}{1+ce^{4x}}$ está <b>acotado</b> entre las soluciones constantes $y_{1}=0$ y $y_{2}=1$, es decir, $0<y(x)<1$ para toda x (subregión $R_{2})$.</li>
<li>El conjunto de soluciones provenientes de la familia uniparamétrica $y(x)=\frac{ce^{4x}}{1+ce^{4x}}$ <b>no cruza</b> ninguna de las soluciones constantes. Este comportamiento es caracteístico de todas las EDO autónomas.</li>
<li>El campo de pendientes en la subregión $R_{2}$ tiene el mismo valor de pendiente ($f(y)>0$), es decir, $f(y)$ <b>no cambia de signo</b>. Este comportamiento se repite en las subregiones $R_{1}$ y $R_{3}$. Esta es otra característica propira de las EDO autónomas.</li>
<li>Puesto que $\frac{dy}{dx}= f(y(x))$ es positiva o negativa en una subregión determinada del gráfico, $y(x)$ es estrictamente <b>monótona</b>, es decir, $y(x)$ está <b> creciendo o decreciendo</b> en una determinada subregión de la gráfica. Por tanto, $y(x)$ <b> no puede oscilar o tener un máximo o un mínimo relativo.</b></li>
</ul>

<h3><span style="color:green"><b>c. Soluciones de equilibrio y diagrama de fase </b></span></h3>

Las soluciones constantes de una EDO autónoma de primer orden son conocidas como soluciones de equilibrio porque ellas <b>no varían a medida que la variable independiente ($x$ en este caso) aumenta o disminuye</b>. 

Las <b>soluciones de equilibrio</b> también se conocen como <b>puntos críticos</b> o puntos estacionarios. Para comprobar que un punto cualquiera es un punto crítico de una EDO autónoma de primer orden, este deberá cumplir que $f(c)=0$.

Por ejemplo, suponga que se necesita resolver la EDO autónoma de primer grado:
$$
\begin{equation}
\begin{aligned}
\frac{dy}{dx}=y(y-3)(y-6)(y-10)
\end{aligned}
\label{ODE_auto_5}
\tag{5}
\end{equation}
$$


Por simple inspección $f(y)=y(y-3)(y-6)(y-10)$. Los puntos críticos, según lo visto anteriormente, se obtienen de $f(y)=0$, es decir, $y(y-3)(y-6)(y-10)=0$. De esta última expresión se desprende que $y_{1}=0$, $y_{2}=3$, $y_{3}=6$ y $y_{4}=10$ son soluciones constantes de la ecuación diferencial planteada. 

Ahora, si se grafica $f(y)$ vs. $y$ se obtiene una gráfica como la que se muestra en la Figura 2:

```python
#Vector of x's
def f(y):
    return y*(y-3)*(y-6)*(y-10)
def g(y):
    return np.zeros(len(y))
y = np.linspace(0,15,100)


plt.plot(y,f(y),color='black')
plt.plot(y,g(y), color = '#bcbd22')
plt.scatter(3,0, color='black')
plt.text(3,0, '({}, {})'.format(3, 0), ha='right',va='bottom')
plt.scatter(6,0, color='black')
plt.text(6.1,0, '({}, {})'.format(6, 0), ha='left',va='bottom')
plt.scatter(10,0, color='black')
plt.text(10.4,0, '({}, {})'.format(10, 0), ha='left',va='bottom')
plt.scatter(10,0, color='black')
plt.text(0.3,-1, '({}, {})'.format(0, 0), ha='left',va='top')
plt.arrow(3,0,-1.4,0,width= 1.8,head_width = 7,head_length=1,color='#bcbd22')
plt.arrow(3,0,1.5,0,width= 1.8,head_width = 7,head_length=1,color='#17becf')
plt.arrow(10,0,-1.5,0,width= 1.8,head_width = 7,head_length=1,color='#bcbd22')
plt.arrow(10,0,2,0,width= 1.8,head_width = 7,head_length=1,color='#17becf')
plt.xlabel('y', size=12)
plt.ylabel('f(y)',size=12)
#plt.legend()
plt.ylim(-190,100)
plt.xlim(0,13)
plt.title('Fig. 2: Gráfica de f(y) vs. y')
plt.show()

```

Para entender las implicaciones de la Figura 2, supongamos que esta describe la dinámica de una partícula. Para ello, asuma que la variable $y$ representa la posición de la partícula y que la velocidad de la misma viene dada por $\frac{dy}{dt}= v$. Esta última ecuación se parece en su forma a la EDO autónoma de primer orden $\frac{dy}{dt}=f(y)$. Por comparación se ve que $f(y)=v$.

Analizando el movimiento de la partícula por intervalos, se pueden obtener las siguientes conclusiones:

<ul>
<li>En el intervalo $y:(0,3)$, $f(y)$ es negativo, es decir, la velocidad de la partícula es negativa y por tanto, la partícula se está dirigiendo hacia la <span style="color:#bcbd22">izquierda</span>.</li>
<li>En el intervalo $y:(3,6)$, $f(y)$ es positivo, es decir, la velocidad de la partícula es positiva y por tanto, la partícula se está dirigiendo hacia la <span style="color:#17becf">derecha</span>.</li>
<li>En el intervalo $y:(6,10)$, $f(y)$ vuelve a ser negativo, es decir, la velocidad de la partícula es negativa y por tanto, la partícula se está dirigiendo hacia la <span style="color:#bcbd22">izquierda</span>.</li>
<li>En el intervalo $y:(10,\infty)$, $f(y)$ vuelve a ser positivo, es decir, la velocidad de la partícula es positiva y por tanto, la partícula se está dirigiendo hacia la <span style="color:#17becf">derecha</span>.</li>
</ul>

<b>¿Y qué pasa en los puntos críticos?</b>

En los puntos en donde $y=0$, $y=3$, $y=6$ y $y=10$, la partícula se encuentra en <b>equilibrio</b>. Esto debido a que en estos puntos $f(y)=0$, es decir, la velocidad de la partícula es 0. Es por esto que estos puntos críticos también se conocen con el nombre de <b>soluciones de equilibrio</b>. 

Estos puntos críticos se encuentran representados en la Figura 2. De especial mención en dicha Figura es la recta de color dorado que contiene las flechas que indican la dirección del movimiento de la partícula. Esta recta es conocida como <b>Diagrama de Fase</b>.

El diagrama de fase (representado de una manera alternativa) y la solución de la EDO autónoma de primer orden ([5](#mjx-eqn-ODE_auto_5)) se muestran en las Figuras 3 y 4 respectivamente:

```python
def f(y,t):
    dydt = y*(y-3)*(y-6)*(y-10)
    return dydt

fig,axs = plt.subplots(1,2,figsize=(11,6))

t= np.linspace(-0.01,0.01,10)

axs[0].vlines(0,-3,13,color='#bcbd22')
axs[0].set_ylim(-3,13)
axs[0].set_xlim(-1,1)
axs[0].scatter(0,0, color='black', s= 80)
axs[0].text(0,0, '({}, {})'.format(0, 0), ha='left',va='bottom',size=14)
axs[0].scatter(0,3, color='black', s= 80)
axs[0].text(0,3, '({}, {})'.format(3, 0), ha='left',va='bottom',size=14)
axs[0].scatter(0,6, color='black')
axs[0].text(0,6, '({}, {})'.format(6, 0), ha='left',va='bottom',size=14)
axs[0].scatter(0,10, color='black')
axs[0].text(0,10, '({}, {})'.format(10, 0), ha='left',va='bottom',size=14)
axs[0].arrow(0,2.5,0,-1,width= 0.03,head_width = 0.1,head_length=0.7,color='#bcbd22')
axs[0].arrow(0,3.5,0,1,width= 0.03,head_width = 0.1,head_length=0.7,color='#17becf')
axs[0].arrow(0,9.5,0,-1,width= 0.03,head_width = 0.1,head_length=0.7,color='#bcbd22')
axs[0].arrow(0,10.5,0,1,width= 0.03,head_width = 0.1,head_length=0.7,color='#17becf')
axs[0].tick_params(axis='both', which='both', labelsize=14, bottom=False, labelbottom=False)
axs[0].set_ylabel(r'y(t)',size=14)
axs[0].set_title('Fig. 3: Diagrama de fase',size=14)

S_values = np.arange(0.0,4,0.005, dtype=float)
vcolors = plt.cm.viridis_r(np.linspace(0,4, len(S_values)))  # colors      for each trajectory
# t points
t = np.linspace(0,0.1)
#Initial condition
y0 = np.arange(-3,10.1,0.2)
for i in range(len(y0)):
    y = odeint(f,y0[i],t)
    axs[1].plot(t,y,label = 'y0 ='+str(y0[i]), color=vcolors[i]) 
axs[1].set_xlabel(r't', size=14)
axs[1].set_ylabel(r'y(t)',size=14)
axs[1].tick_params(axis='both', which='major', labelsize=14)
axs[1].set_ylim(-3,13)
axs[1].set_title('Fig. 4: Solución de la EDO autónoma 5',size=14)
plt.show()
```

<h3><span style="color:green"><b>d. Análisis cualitativo. Tipos de puntos críticos </b></span></h3>

<h4><span style="color:#d07200"><i>d.1 Puntos críticos inestables y asintóticamente estables</i></span></h4>

Del diagrama de fase y de la solución de la ecuación diferencial podemos obtener valiosa información sobre el comportamiento cualitativo de las soluciones de la EDO autónoma de primer orden ([5](#mjx-eqn-ODE_auto_5)). Las gráficas de las soluciones $y_{1}=0$, $y_{2}=3$, $y_{3}=6$ y $y_{4}=10$ (Figura 4) son líneas rectas paralelas al eje $t$. Esto quiere decir que su valor no cambia con el aumento o disminución de $t$, motivo por el cual reciben el nombre de <b> soluciones constantes</b>

Tome como ejemplo el punto (0,3). Al observar la Figura 4, se puede notar que las curvas solución que pasan a través de puntos lo suficientemente cercanos a ambos lados de la solución constante $y_{2}=3$, tienden a alejarse a medida que $t\rightarrow\infty$.A este tipo de puntos críticos en donde las curvas solución tienden a alejarse a medida que $t\rightarrow\infty$ se los conoce como <b>inestables</b>. Una manera fácil de reconocerlos es mediante el diagrama de fase: las flechas del diagrama <b>divergen</b> de un punto crítico inestable (Figura 3), razón por la cual también se lo conoce con el nombre de <b>repulsor</b>.

Tome como ejemplo el punto (0,6). Al observar la Figura 4, se puede notar que las curvas solución que pasan a través de puntos lo suficientemente cercanos a ambos lados de la solución constante $y=6$, tienden a acercarse a medida que $t\rightarrow\infty$. A este tipo de puntos críticos en donde las curvas solución tienden a acercarse a medida que $t\rightarrow\infty$ (acercamiento asintótico) se los conoce como <b>asintóticamente estables</b>. Una manera fácil de reconocerlos es mediante el diagrama de fase: las flechas del diagrama <b>convergen</b> de un punto crítico asintóticamente estable (Figura 3), razón por la cual también se lo conoce con el nombre de <b>atractor</b>.

<h4><span style="color:#d07200"><i>d.2 Puntos críticos semiestables</i></span></h4>

Considere la siguiente ecuación diferencial:

$$
\begin{equation}
\begin{aligned}
\frac{dy}{dt}= y^{2}
\end{aligned}
\label{ODE_auto_6}
\tag{6}
\end{equation}
$$

Por simple inspección $f(y)=y^{2}$. Los puntos críticos, según lo visto anteriormente, se obtienen de $f(y)=0$, es decir, $y^{2}=0$. De esta última expresión se desprende que $y_{1}=0$ y $y_{2}=0$ son soluciones constantes de la ecuación diferencial planteada. 

```python
fig,axs = plt.subplots(1,2,figsize=(10,5))

def g(y):
    return y**2
y = np.linspace(-4,4)
axs[0].plot(y,g(y))
axs[0].scatter(0,0, color='black')
axs[0].text(0,1, '({}, {})'.format(0,0), ha='center',va='top')
axs[0].set_title('Fig. 5: Gráfica de f(y) vs y')
axs[0].hlines(0,-4,4,color='#1f77b4')
axs[0].arrow(1,0,1.4,0,width= 0.3,head_width = 0.8,head_length=1,color='#bcbd22')
axs[0].arrow(-3,0,1.5,0,width= 0.3,head_width = 0.8,head_length=1,color='#17becf')
axs[0].set_xlabel(r'y',size=14)
axs[0].set_ylabel(r'f(y)',size=14)


def f(Y,X):
    return Y**2
X,Y = np.meshgrid(np.linspace(-4,4,15),np.linspace(-4,4,15))
#vectorizing the color map
S_values = np.arange(0.0,2.25,0.25, dtype=float)
vcolors = plt.cm.viridis_r(np.linspace(0.5,1, len(S_values)))  # colors      for each trajectory
U,V = 1,f(Y,X)
N = np.sqrt(U**2 + V**2)
U /= N
V /= N
axs[1].quiver(X, Y, U, V ,color = vcolors,scale=20,linewidth=0.05)
axs[1].set_title('Fig. 6: Campo de direcciones de la EDO autónoma 5')
axs[1].set_xlabel(r't',size=14)
axs[1].set_ylabel(r'y(t)',size=14)
plt.show()
```

En la Figura 5 se observa que el punto (0,0) es el único punto crítico de la EDO autónoma de primer orden ([6](#mjx-eqn-ODE_auto_6)). Si realizamos nuevamente la analogía con el mundo físico, en el intervalo $y:(-\infty,0)$, $f(y)>0$ (la velocidad de la partícula aumenta), por lo que la partícula estará dirigiéndose a la derecha. Similarmente, en el intervalo $y:(0,\infty)$, $f(y)>0$ (nuevamente la velocidad de la partícula se incrementa), por lo que la partícula estará dirigiéndose a la derecha. En el punto crítico, la partícula se encuentra en equilibrio. 

En la Figura 6 se puede observar el campo de direcciones de la EDO autónoma de primer orden ([6](#mjx-eqn-ODE_auto_6)). Los vectores se orientan verticalmente, es decir,$f(y)>0$ con pendiente tendiendo al infinito para valores en donde $y\rightarrow\pm\infty$. Los vectores que están cercanos a la solución constante $y=0$ cambian su pendiente, hasta estar completamente horizontales en $y=0$ (en este punto $f(y)=0$).

```python
fig,axs = plt.subplots(1,2,figsize=(10,5))

axs[0].arrow(0,1,0,1,width= 0.09,head_width = 0.2,head_length=1,color='#bcbd22')
axs[0].arrow(0,-2.5,0,1,width= 0.09,head_width = 0.2,head_length=1,color='#17becf')
axs[0].scatter(0,0, color='black')
axs[0].text(0.1,0, '({}, {})'.format(0, 0), ha='left',va='bottom',size=14)
axs[0].vlines(0,-4,4,color='#bcbd22')
axs[0].tick_params(axis='both', which='both', labelsize=14, bottom=False, labelbottom=False)
axs[0].set_ylabel(r'y(t)',size=14)
axs[0].set_title('Fig. 7: Diagrama de fase',size=14)
axs[0].set_ylim(-4,4)
axs[0].set_xlim(-1,1)


def f(t,c):
     return -1/(t+c)
x = np.linspace(-4,4,200)
c = [-3,-2,-1,0,1,2,3]

#vectorizing the color map
S_values = np.arange(0.0,2.25,0.25, dtype=float)
vcolors = plt.cm.viridis_r(np.linspace(0.5,1, len(S_values)))  # colors      for each trajectory

for i in range(len(c)):
    y = [f(j,c[i]) for j in x]
    for k in range(len(y)):
        if (y[k]-y[k-1]<0):
            y[k]= np.nan
            axs[1].plot(x,y, label = 'c='+ str(c[i]), color=vcolors[i]) 
axs[1].set_xlabel('t', size=12)
axs[1].set_ylabel('y(t)',size=12)
axs[1].legend()
axs[1].set_ylim(-4,4)
axs[1].set_title(r'$Fig.8:~Familia~de~soluciones~y= \frac{-1}{t+c},~~y = 0$', size=14)
axs[1].hlines(0,-4,4,color='#bcbd22')
plt.show()
```

En la Figura 8 se observa la familia de soluciones de la EDO autónoma de primer orden ([6](#mjx-eqn-ODE_auto_6)). Las curvas solución de color púrpura-azul corresponden a la familia de soluciones $y = \frac{-1}{t+c}$, mientras que la curva de color dorado corresponde a la solución singular $y=0$. 

Cotejando con el campo de direcciones de la Figura 6, se puede notar que por un lado, las curvas solución que pasan por puntos lo suficientemente cercanos a la solución constante $y=0$, tienden a $y=0$ a medida que $t\rightarrow\infty$, en el intervalo $y:(-\infty,0)$. Por otra parte, en el intervalo $y:(0,\infty)$, las curvas solución que pasan por puntos lo suficientemente cercanos a la solución constante $y=0$, se alejan de la misma a medida que $t\rightarrow\infty$. Este tipo de puntos críticos que presentan un comportamiento como el detallado anteriormente se los conoce con el nombre de <b>semiestables</b>. 

Una manera fácil de reconocerlos es mediante el diagrama de fase: para un punto crítico semiestable, las flechas del diagrama no convergen ni divergen (apuntan siempre en la misma dirección)(Figura 7).

<h3><span style="color:green"><b>e. Linearización alrededor de un punto crítico </b></span></h3>

Suponga que $y_{1}$ es un punto crítico de la EDO autónoma $\frac{dy}{dx}=f(y)$ y asuma que f(y) tiene una derivada continua en la vencindad de $y_{1}$:

<ul>
<li>Si $\frac{d}{dy}f(y_{1})<0$, entonces $y_{1}$ es un punto crítico <b>asintóticamente estable</b>.</li>
<li>Si $\frac{d}{dy}f(y_{1})>0$, entonces $y_{1}$ es un punto crítico <b>inestable</b>.</li>
<li>Si $\frac{d}{dy}f(y_{1})=0$, entonces se necesita más información para determinar la naturaleza de $y_{1}$.</li>
</ul>

<h1><span style="color:purple">2.3 Ecuaciones lineales de primer orden<span></h1>

<h2><span style="color:blue" >2.3.1 Definición<span></h2>

Uns ecuación diferencial de la forma:

$$
\begin{equation}
\begin{aligned}
a_{1}(x)\frac{dy}{dx}+a_{0}(x)y = g(x)
\end{aligned}
\label{ODE_FO_1}
\tag{7}
\end{equation}
$$

se dice que es una ecuación lineal en la variable dependiente y.

<h2><span style="color:blue" >2.3.2 Forma estándar de una EDO lineal de primer orden<span></h2>

Al dividir ambos miembros de la EDO lineal ([7](#mjx-eqn-ODE_FO_1)) para $a_{1}(x)$ se obtienen la forma estándar de una EDO lienal de primer orden:
$$
\begin{equation}
\begin{aligned}
\frac{dy}{dx}+P(x)y &= f(x)\\
P(x) &= \frac{a_{0}(x)}{a_{1}(x)}\\
f(x) &= \frac{g(x)}{a_{1}(x)}
\end{aligned}
\label{ODE_FO_2}
\tag{8}
\end{equation}
$$

De la forma estándar ([8](#mjx-eqn-ODE_FO_2)) se puede notar que $a_{1}(x)\neq 0$ para garantizar la existencia de $P(x)$ y $f(x)$. De esto se desprende que se debe buscar una solución de la EDO ([8](#mjx-eqn-ODE_FO_2)) para la cual $P(x)$ y $f(x)$ sean <b>funciones continuas</b>. 

<h2><span style="color:blue" >2.3.3 Algoritmo de solución<span></h2>

Para resolver una EDO lineal de primer orden, se deben seguir el siguiente algoritmo: 

<span style="color:green" ><b>a. Colocar la EDO lineal en su forma estándar</b></span>

Por ejemplo, considere la siguiente EDO lineal:

$$
\begin{equation}
\begin{aligned}
x\frac{dy}{dx}-y= x^{2}\sin{x}
\end{aligned}
\label{ODE_FO_3}
\tag{9}
\end{equation}
$$

Dividiendo ambos miembros de la EDO lineal ([9](#mjx-eqn-ODE_FO_3)) entre $a_{1}(x)=x$, se obtiene su forma estándar:

$$
\begin{equation}
\begin{aligned}
\frac{dy}{dx}-\frac{y}{x} = x\sin{x}
\end{aligned}
\label{ODE_FO_4}
\tag{10}
\end{equation}
$$

De la ecuación ([10](#mjx-eqn-ODE_FO_4)), se puede identificar que $P(x)=-\frac{1}{x}$ y $f(x)=x\sin{x}$. Debe observarse que $P(x)$ no es una función continua puesto que $x=0$ produce una indeterminación.

<span style="color:green"><b>b. Encontrar el factor de integración</b></span>

Para encontrar el factor de integración de una EDO lineal de primer orden, multiplique ambos miembros de la ecuación ([8](#mjx-eqn-ODE_FO_2)) por un factor $\mu(x)$:

$$
\begin{equation}
\begin{aligned}
\mu(x)\frac{dy}{dx}+\mu(x)P(x)y &= \mu(x) f(x)\\
\end{aligned}
\label{ODE_FO_5}
\tag{11}
\end{equation}
$$

Ahora, observe que:

$$
\begin{equation}
\begin{aligned}
\frac{d}{dx}\left[\mu(x)y\right]= \mu(x)\frac{dy}{dx}+\frac{d\mu(x)}{dx}y
\end{aligned}
\label{ODE_FO_6}
\tag{12}
\end{equation}
$$

Comparando las ecuaciones ([11](#mjx-eqn-ODE_FO_5)) y ([12](#mjx-eqn-ODE_FO_6)):

$$
\begin{align*}
\frac{d\mu(x)}{dx}\color{red}{y}&= \mu(x)P(x)\color{red}{y}\\
\frac{d\mu(x)}{dx} &= \mu(x)P(x)\\
\frac{d\mu(x)}{\mu(x)} &= P(x)dx\\
\int\frac{d\mu(x)}{\mu(x)} &= \int P(x)dx\\
\ln{|\mu(x)|} &=\int P(x)dx + C\\
|\mu(x)| &= c_{1}e^{\int P(x)dx }~~\Rightarrow~~ \color{#1f77b4}{c_{1}}= \color{#1f77b4}{e^{c}}\\
\mu(x) &= c_{2}e^{\int P(x)dx }~~\Rightarrow~~ \color{#1f77b4}{c_{2}}= \color{#1f77b4}{\pm c_{1}}
\end{align*}
$$

En la última ecuación, la constante $c_{2}$ puede tomar cualquier valor, y todos ellos producirán el mismo resultado. Por tanto, se tomará el valor de $c_{2}=1$:

$$
\begin{equation}
\begin{aligned}
\mu(x) &= e^{\int P(x)dx }
\end{aligned}
\label{ODE_FO_7}
\tag{13}
\end{equation}
$$

La ecuación ([13](#mjx-eqn-ODE_FO_7)) permite determinar el valor del factor de integración de una EDO lineal en su forma estándar. 

Aplicando esto a la EDO lineal  ([10](#mjx-eqn-ODE_FO_4)):
$$
\begin{align*}
\mu(x) &= e^{\int -\frac{1}{x} dx}\\
\mu(x) &= e^{-\ln{|x|}}\\
\mu(x) &= e^{\ln{|x|^{-1}}}\\
\mu(x) &= |x|^{-1}\\
\mu(x) &= \left|\frac{1}{x}\right|
\end{align*}
$$

De la expresión es notable que $x\neq0$. Luego, por la definición de valor absoluto:
$$
\begin{align*}
\left|\frac{1}{x}\right|= \left\{
\begin{array}{ll}
\frac{1}{x} & ,\frac{1}{x}\geq 0\\
-\frac{1}{x} & ,\frac{1}{x}<0\\
\end{array}
\right.\\
\end{align*}
$$

Eligiendo el intervalo de definición de la solución de la EDO ([10](#mjx-eqn-ODE_FO_4)) como  $0<x<\infty$, permite seleccionar a $\mu(x) = \frac{1}{x}$ como el factor de integración. Luego, multiplicando este factor en ambos miembros de la EDO ([10](#mjx-eqn-ODE_FO_4)):

$$
\begin{align*}
\frac{1}{x}\frac{dy}{dx}-\frac{1}{x}\frac{y}{x} &= \frac{1}{x}x\sin{x}\\
\color{#1f77b4}{\frac{1}{x}\frac{dy}{dx}-\frac{1}{x^{2}}y} &= \sin{x}
\end{align*}
$$

El lado izquierdo de la última expresión proviene de:
$$
\begin{align*}
\frac{d}{dx}\left[\frac{1}{x}y\right]= \color{#1f77b4}{\frac{1}{x}\frac{dy}{dx}-\frac{1}{x^{2}}y}
\end{align*}
$$

Por tanto:
$$
\begin{align*}
\int\frac{d}{dx}\left[\frac{1}{x}y\right]dx=\int\sin{x}dx
\end{align*}
$$

Obtieniendo:
$$
\begin{align*}
\frac{1}{x}y &= -\cos{x}+c\\
y &= -x\cos{x}+ xc
\end{align*}
$$


La solución se encuentra definida en el intervalo $x:(0,\infty)$. 

```python
def f(x,c):
    return -x*np.cos(x) + x*c

x = np.linspace(0,4,500)
#c = np.linspace(-5,5,1)
c = [-4,-2,-2,0,1,2,4]
#vectorizing the color map
S_values = np.arange(0.0,10,0.25, dtype=float)
vcolors1 = plt.cm.viridis_r(np.linspace(0,0.5, len(S_values))) 
vcolors2 = plt.cm.viridis_r(np.linspace(0,10, len(S_values))) 
for i in range(len(c)):
    y = [f(j,c[i]) for j in x]
    plt.plot(x,y,label = 'c='+ str(c[i]), color=vcolors2[i])
plt.legend()
#plt.xlim(-1.5,1.5)
#plt.ylim(-0.5,1.5)
plt.xlabel('x', size=14)
plt.ylabel('y(x)', size=14)
plt.title ('Fig. 9: Familia de soluciones de la EDO lineal 10')
plt.show()
```

<h2><span style="color:blue" >2.3.4 Solución general<span></h2>

De la ecuación ([11](#mjx-eqn-ODE_FO_5)) y aplicando el procedimiento visto anteriormente:
$$
\begin{align*}
\color{#1f77b4}{\mu(x)}\frac{dy}{dx}+\color{#1f77b4}{\mu(x)}P(x)y &= \color{#1f77b4}{\mu(x)} f(x)\\
\color{#1f77b4}{e^{\int P(x)dx}}\frac{dy}{dx}+\color{#1f77b4}{e^{\int P(x)dx}}P(x)y &= \color{#1f77b4}{e^{\int P(x)dx}} f(x)\\
\end{align*}
$$

El lado izquierdo de la última expresión proviene de:
$$
\begin{align*}
\frac{d}{dx}\left[e^{\int P(x)dx}y\right]= \color{#1f77b4}{e^{\int P(x)dx}}\frac{dy}{dx}+\color{#1f77b4}{e^{\int P(x)dx}}P(x)y 
\end{align*}
$$

Por tanto:
$$
\begin{align*}
\int\frac{d}{dx}\left[e^{\int P(x)dx}y\right]dx= \int e^{\int P(x)dx}f(x)dx
\end{align*}
$$

De donde:
$$
\begin{align*}
e^{\int P(x)dx}y= \int e^{\int P(x)dx}f(x)dx + c
\end{align*}
$$  

Resultando:

$$
\begin{equation}
\begin{aligned}
y= e^{-\int P(x)dx} \int e^{\int P(x)dx}f(x)dx + ce^{-\int P(x)dx}
\end{aligned}
\label{ODE_FO_8}
\tag{14}
\end{equation}
$$

Bajo la suposición de que $P(x)$ y $f(x)$ son continuas en un intervalo común $I$, la solución en $I$ de una EDO lineal de primer orden $\frac{dy}{dx} + P(x)y = f(x)$ viene dada por una función que tiene la forma que muestra la ecuación ([14](#mjx-eqn-ODE_FO_8)). Recíprocamente, es posible demostrar que cualquier función de la forma que presenta la ecuación ([14](#mjx-eqn-ODE_FO_8)) es una solución de la EDO lineal de primer orden $\frac{dy}{dx} + P(x)y = f(x)$ sobre $I$.  

Esto quiere decir que la ecuación ([14](#mjx-eqn-ODE_FO_8)) es una familia uniparamétrica de soluciones de $\frac{dy}{dx} + P(x)y = f(x)$ y <b>toda solución de esta EDO en I es miembro de esta familia.</b> Por tanto, la ecuación ([14](#mjx-eqn-ODE_FO_8)) es la <b>solución general</b> de la ecuación diferencial $\frac{dy}{dx} + P(x)y = f(x)$.

<h2><span style="color:blue" >2.3.5 Problema de valor inicial<span></h2>

Considere una EDO lineal en su forma estándar (ecuación ([8](#mjx-eqn-ODE_FO_2))):

$$
\begin{align*}
\frac{dy}{dx}+P(x)y &= f(x)\\
\end{align*}
$$

sujeta a la condición inicial $y(x_{0})=y_{0}$.

Esta ecuación puede escribirse en la forma:
$$
\begin{equation}
\begin{aligned}
\frac{dy}{dx} &= F(x,y)\\
\frac{dy}{dx} &= -P(x)y+f(x)
\end{aligned}
\label{ODE_FO_9}
\tag{15}
\end{equation}
$$

De la ecuación  ([15](#mjx-eqn-ODE_FO_9)) se puede reconocer que $F(x,y)=-P(x)y+f(x)$. Dado que $P(x)$ y $f(x)$ son continuas sobre un intervalo $I$, $F(x,y)$ y su derivada $\frac{\partial F}{\partial y}$ también serán continuas sobre el mismo intervalo. Con esto, basados en el <b>Teorema de existencia y unicidad de una solución para una EDO de primer orden</b>, podemos afirmar que existe un solución para el PVI y esta es única. Esta estará definida en un intervalo $I_{0}$ que contiene a $x_{0}$.

Nótese que $I$ e $I_{0}$ no son iguales. El primero es el intervalo de definición de la solución general y el segundo es el intervalo de definición del PVI (solución particular).

Considere por ejemplo la ecuación diferencial:
$$
\begin{equation}
\begin{aligned}
\frac{dy}{dx}+ y\tan{x} = \cos^2{x}
\end{aligned}
\label{ODE_FO_10}
\tag{16}
\end{equation}
$$

sujeta a la condición inicial $y(0)=-1$.

Lo primero que tiene que observarse de la ecuación ([16](#mjx-eqn-ODE_FO_10)) es que la función $\tan{x}$ no es continua, es decir, $P(x)$ no es continua. Para entender esto, observe la Figura 10:

```python
x = np.arange(-2*np.pi,2*np.pi,0.1)
y = np.tan(x)
y[:-1][np.diff(y) < 0] = np.nan
plt.vlines(-np.pi/2,-2,2,linestyles='--', color = '#ff7f0e')
plt.vlines(np.pi/2,-2,2,linestyles='--', color = '#ff7f0e')
plt.text(-np.pi/2,1.5,r'$x=-\frac{\pi}{2}$',size = 14)
plt.text(np.pi/2,1.5,r'$x=\frac{\pi}{2}$',size = 14)
plt.plot(x,y)        
#plt.legend()
#plt.xlim(-1.5,1.5)
plt.ylim(-2,2)
plt.xlabel('x', size=14)
plt.ylabel('tan(x)', size=14)
plt.title ('Fig. 10: Gráfica de P(x)=tan(x)')
plt.show()
```

La función $\tan{x}$ solo es continua en algunos intervalos, como por ejemplo $x:(-\frac{\pi}{2},\frac{\pi}{2})$, intervalo demarcado por las líneas interpuntadas en la Figura 10. Por tanto, la solución de la EDO ([16](#mjx-eqn-ODE_FO_10)) tendrá que estar enmarcada en este intervalo.

Calculando el factor de integración se tiene:
$$
\begin{align*}
\mu(x) &= e^{\int \tan{x}~dx}\\
\mu(x) &= e^{-\ln{|\cos{x}|}}\\
\mu(x) &= e^{\ln{|\cos{x}|^{-1}}}\\
\mu(x) &= |\cos{x}|^{-1}\\
\mu(x) &= \left|\sec{x}\right|
\end{align*}
$$


Tome el valor positivo del valor absoluto, y defina $\mu(x)= \sec{x}$. Multiplicando ambos miembros de la EDO ([16](#mjx-eqn-ODE_FO_10)) por el factor de integración se tiene:
$$
\begin{align*}
\sec{x}\frac{dy}{dx}+ y\sec{x}\tan{x} &= \sec{x}\cos^2{x}\\
\color{#1f77b4}{\sec{x}\frac{dy}{dx}+ y\sec{x}\tan{x}} &= \cos{x}\\
\frac{d}{dx}\left[\sec{x}~y\right] &= \color{#1f77b4}{\sec{x}\frac{dy}{dx}+ y\sec{x}\tan{x}}
\end{align*}
$$

Por tanto:
$$
\begin{align*}
\int \frac{d}{dx}\left[\sec{x}~y\right] &= \int \cos{x}\\
sec{x}~y &= \sin{x} + c\\
y &= \sin{x}\cos{x} + c \cos{x}
\end{align*}
$$

Haciendo cumplir la condición inicial $(0,-1)$:
$$
\begin{align*}
y &= \sin{x}\cos{x} + c\cos{x}\\
-1 &= \sin{0}\cos{0} + c\cos{0}\\
-1 &= c\\
y &= \sin{x}\cos{x} -\cos{x}
\end{align*}
$$

Las gráficas de la familia de soluciones de la EDO ([16](#mjx-eqn-ODE_FO_10)) y de la solución particular se muestran en la Figura 11:

```python
def f(x,c):
    return np.sin(x)*np.cos(x)+c*np.cos(x)
x = np.linspace(-2*np.pi,2*np.pi,100)
c = [-2,-1,0,1,2]
plt.vlines(-np.pi/2,-2,2,linestyles='--', color = 'black')
plt.text(-np.pi/2,1.5,r'$x=-\frac{\pi}{2}$',size = 14)
plt.text(np.pi/2,1.5,r'$x=\frac{\pi}{2}$',size = 14)
plt.vlines(np.pi/2,-2,2,linestyles='--', color = 'black')
plt.scatter(0,-1, color='black')
plt.text(-1,-0.8, '({}, {})'.format(0,-1), ha='left',va='bottom',size=14)

#vectorizing the color map
S_values = np.arange(0.0,10,0.25, dtype=float)
vcolors1 = plt.cm.viridis_r(np.linspace(0,0.5, len(S_values))) 
vcolors2 = plt.cm.viridis_r(np.linspace(0.2,4.5, len(S_values))) 

for i in range(len(c)):
    y = [f(j,c[i]) for j in x]
    plt.plot(x,y,label = 'c='+ str(c[i]), color=vcolors2[i])
#plt.legend()
#plt.xlim(-1.5,1.5)
plt.ylim(-2,2)
plt.xlabel('x', size=14)
plt.ylabel('y(x)', size=14)
plt.title ('Fig. 11: Solución particular de la EDO 10 en (0,-1)')
plt.show()
```

Esta solución estará definida en $x:(-\frac{\pi}{2},\frac{\pi}{2})$.

<h1><span style="color:purple">2.4 Ecuaciones exactas<span></h1>

<h2><span style="color:blue" >2.4.1 Diferencial de una función de dos variables<span></h2>

Sea $f(x,y)$ una función de dos variables, cuyas primeras derivadas $\frac{\partial f}{\partial x}$ y $\frac{\partial f}{\partial y}$ son continuas en una región $R$ del plano $xy$. Entonces, su diferencial viene dado por: 
$$
\begin{equation}
\begin{aligned}
df=\left(\frac{\partial f}{\partial x}\right)_{y}dx+\left(\frac{\partial f}{\partial y}\right)_{x}dy
\end{aligned}
\label{ODE_EX_1}
\tag{17}
\end{equation}
$$

En el caso que $f(x,y)=c$, en donde $c$ es una constante, entonces su diferencial $df=0$, lo cual implica:
$$
\begin{equation}
\begin{aligned}
\left(\frac{\partial f}{\partial x}\right)_{y}dx+\left(\frac{\partial f}{\partial y}\right)_{x}dy=0
\end{aligned}
\label{ODE_EX_2}
\tag{18}
\end{equation}
$$


<h2><span style="color:blue" >2.4.2 Definición de ecuación exacta<span></h2>

En la ecuación ([17](#mjx-eqn-ODE_EX_1)), sea:
$$
\begin{equation}
\begin{aligned}
\left(\frac{\partial f}{\partial x}\right)_{y} &= M(x,y)\\
\left(\frac{\partial f}{\partial y}\right)_{x} &= N(x,y)
\end{aligned}
\label{ODE_EX_3}
\tag{19}
\end{equation}
$$

Entonces, una expresión diferencial $M(x,y)dx+N(x,y)dy$ es una diferencial exacta en una región $R$ del plano $xy$ si esta corresponde a la diferencial de alguna función f(x,y) definida en R. 

Una ecuación diferencial de primer orden de la forma:
$$
\begin{equation}
\begin{aligned}
M(x,y)dx+N(x,y)dy=0
\end{aligned}
\label{ODE_EX_4}
\tag{20}
\end{equation}
$$

se conoce como una <b>ecuación exacta</b> si el lado izquierdo de la ecuación es un diferencial exacto.

<h2><span style="color:blue" >2.4.3 Criterio para una diferencial exacta<span></h2>

Si $M(x,y)$ y $N(x,y)$ son <b>funciones continuas</b> y tienen <b>primeras derivadas parciales continuas</b> en una región rectangular $R$ definida por $a<x<b$ y $c<y<d$, entonces, una condición <b>necesaria y suficiente</b> para que $M(x,y)dx+N(x,y)dy$ sea una diferencial exacta es:

$$
\begin{equation}
\begin{aligned}
\frac{\partial M}{\partial y}= \frac{\partial N}{\partial x}
\end{aligned}
\label{ODE_EX_5}
\tag{21}
\end{equation}
$$

<h2><span style="color:blue" >2.4.4 Algoritmo de solución<span></h2>

<h4><span style="color:#d07200"><i>a. Verificación del criterio de diferencial exacto</i></span></h4>

Dada la ecuación diferencial $M(x,y)dx+N(x,y)dy=0$ (ecuación ([20](#mjx-eqn-ODE_EX_4))), verifique que se cumpla el criterio para una diferencial exacta dado por la ecuación ([21](#mjx-eqn-ODE_EX_5)).

<h4><span style="color:#d07200"><i>b. Planteamiento</i></span></h4>

Si se cumple el criterio anterior, entonces existe una función $f$ para la cual se cumple la ecuación ([19](#mjx-eqn-ODE_EX_3))) :
$$
\begin{align*}
\left(\frac{\partial f}{\partial x}\right)_{y}= M(x,y)~~~~~\left(\frac{\partial f}{\partial y}\right)_{x}= N(x,y)
\end{align*}
$$
    

Tómese la expresión de mano derecha. Es posible calcular $f$ integrando $N(x,y)$ con respecto a y, manteniendo $x$ constante. Al hacer esto, la derivada parcial con respecto a $y$ se transforma en una derivada total:
$$
\begin{align*}
\left(\frac{\partial f}{\partial y}\right)_{x}= N(x,y)~~\Rightarrow~~si~~x~~es~~constante:~~\frac{df}{dy}=N(x,y)
\end{align*}
$$

Entonces:
$$
\begin{equation}
\begin{aligned}
f(x,y) = \int N(x,y) dy + \color{green}{h(x)}
\end{aligned}
\label{ODE_EX_6}
\tag{22}
\end{equation}
$$


donde la función <b>arbitraria</b> $h(x)$ es la "constante" de integración. Ahora, tomando la derivada parcial de la ecuación  ([22](#mjx-eqn-ODE_EX_6)) con respecto a x:


$$
\begin{align*}
\left[\frac{\partial f(x,y)}{\partial x}\right]_{y}= \frac{\partial}{\partial x}\left[\int N(x,y) dy + \color{green}{h(x)}\right]= \frac{\partial}{\partial x}\int N(x,y) dy + \color{blue}{h'(x)}
\end{align*}
$$

Recordando que $\left(\frac{\partial f}{\partial x}\right)_{y}= M(x,y)$, entonces:

$$
\begin{align*}
\frac{\partial}{\partial x}\int N(x,y) dy + \color{blue}{h'(x)} = M(x,y)
\end{align*}
$$

Despejando $h'(x)$:

$$
\begin{equation}
\begin{aligned}
\color{blue}{h'(x)} = M(x,y)-\frac{\partial}{\partial x}\int N(x,y) dy
\end{aligned}
\label{ODE_EX_7}
\tag{23}
\end{equation}
$$

La ecuación ([23](#mjx-eqn-ODE_EX_7)) se integra con respecto a $x$ y se sustituye el resultado en la ecuación ([22](#mjx-eqn-ODE_EX_6)). La solución implícita de la EDO de primer orden exacta ([20](#mjx-eqn-ODE_EX_4)) es $f(x,y)=c$.

El algorito de solución puede comenzar también con $\left(\frac{\partial f}{\partial x}\right)_{y}= M(x,y)$. En este caso:

$$
\begin{align*}
\left(\frac{\partial f}{\partial x}\right)_{y}= M(x,y)~~\Rightarrow~~si~~y~~es~~constante:~~\frac{df}{dx}=M(x,y)
\end{align*}
$$

$$
\begin{equation}
\begin{aligned}
f(x,y) = \int M(x,y) dx + \color{green}{g(y)}
\end{aligned}
\label{ODE_EX_8}
\tag{24}
\end{equation}
$$

donde la función arbitraria $g(y)$  es la "constante" de integración. Ahora, tomando la derivada parcial de la ecuación ([24](#mjx-eqn-ODE_EX_8)) con respecto a y:

$$
\begin{align*}
\left[\frac{\partial f(x,y)}{\partial y}\right]_{x} = \frac{\partial}{\partial y}\left[\int M(x,y) dx + \color{green}{g(y)}\right]= \frac{\partial}{\partial y}\int M(x,y) dx + \color{blue}{g'(y)}
\end{align*}
$$

Recordando que $\left(\frac{\partial f}{\partial y}\right)_{x}= N(x,y)$, entonces:

$$
\begin{align*}
\frac{\partial}{\partial y}\int M(x,y) dx + \color{blue}{g'(y)} = N(x,y)
\end{align*}
$$

Despejando $g'(y)$:

$$
\begin{equation}
\begin{aligned}
\color{blue}{g'(y)} = N(x,y)-\frac{\partial}{\partial y}\int M(x,y) dx
\end{aligned}
\label{ODE_EX_9}
\tag{25}
\end{equation}
$$

La ecuación ([25](#mjx-eqn-ODE_EX_9)) se integra con respecto a $y$ y se sustituye el resultado en la ecuación ([24](#mjx-eqn-ODE_EX_8)). La solución implícita de la EDO de primer orden exacta ([20](#mjx-eqn-ODE_EX_4)) es $f(x,y)=c$, igual que en el caso anterior.

<h2><span style="color:blue" >2.4.5 Ejemplos de ecuaciones exactas<span></h2>

Como ejemplo, considere la siguiente ecuación diferencial:
$$
\begin{equation}
\begin{aligned}
\left(y\ln{y}-e^{-xy}\right)dx+\left(\frac{1}{y}+x\ln{y}\right)dy=0
\end{aligned}
\label{ODE_EX_10}
\tag{26}
\end{equation}
$$


Para comprobar que el lado izquierdo de la ecuación ([26](#mjx-eqn-ODE_EX_10)) es un diferencial exacto, aplicamos el criterio dado por la ecuación ([21](#mjx-eqn-ODE_EX_5)):

$$
\begin{align*}
M(x,y) &= y\ln{y}-e^{-xy}\\
\frac{\partial M}{\partial y} &= \frac{\partial y}{\partial y}\ln{y}+y\frac{\partial}{\partial y}\ln{y}-\frac{\partial}{\partial y}e^{-xy}\\
\frac{\partial M}{\partial y} &= \ln{y}+y\frac{1}{y}-xe^{xy}\ln{e}\\
\frac{\partial M}{\partial y} &= \color{green}{1+\ln{y}-xe^{xy}}
\end{align*}
$$

$$
\begin{align*}
N(x,y) &= \frac{1}{y}+x\ln{y}\\
\frac{\partial N}{\partial x} &= 0 + \ln{y}\\
\frac{\partial N}{\partial x} &= \color{green}{\ln{y}}
\end{align*}
$$

De lo anterior se desprende que $\frac{\partial M}{\partial y}\neq\frac{\partial N}{\partial x}$. Al no cumplirse el criterio dado por la ecuación ([21](#mjx-eqn-ODE_EX_5)), se concluye que el lado izquierdo de la ecuación ([26](#mjx-eqn-ODE_EX_10)) no es un diferencial exacto, y por tanto, la ecuación ([26](#mjx-eqn-ODE_EX_10)) no es exacta. 

Considere ahora la siguiente ecuación diferencial:
$$
\begin{equation}
\begin{aligned}
(\sin{y}-y\sin{x})dx+(\cos{x}+x\cos{y}-y)dy=0
\end{aligned}
\label{ODE_EX_11}
\tag{27}
\end{equation}
$$

Para comprobar que el lado izquierdo de la ecuación ([27](#mjx-eqn-ODE_EX_11)) es un diferencial exacto, aplicamos el criterio dado por la ecuación ([21](#mjx-eqn-ODE_EX_5)):

$$
\begin{align*}
M(x,y) &= \sin{y}-y\sin{x}\\
\frac{\partial M}{\partial y} &= \frac{\partial}{\partial y}(\sin{y})-\frac{\partial}{\partial y}(y\sin{x})\\
\frac{\partial M}{\partial y} &= \cos{y}-\sin{x}
\end{align*}
$$

$$
\begin{align*}
N(x,y) &= \cos{x}+x\cos{y}-y\\
\frac{\partial N}{\partial x} &= \frac{\partial}{\partial x}(\cos{x})+\frac{\partial}{\partial x}(x\cos{y})-\frac{\partial}{\partial x}(y)\\
\frac{\partial N}{\partial x} &= -\sin{x} + \cos{y}
\end{align*}
$$

De esto se concluye que $\frac{\partial M}{\partial y}=\frac{\partial N}{\partial x}$, cumpliéndose el criterio dado por la ecuación ([21](#mjx-eqn-ODE_EX_5)). Por tanto, el lado izquierdo de la ecuación ([27](#mjx-eqn-ODE_EX_11)) es un diferencial exacto, y entoces la ecuación ([27](#mjx-eqn-ODE_EX_11)) es exacta. 


$$
\begin{align*}
\left(\frac{\partial f}{\partial x}\right)_{y}=\sin{y}-y\sin{x}~~\Rightarrow~~si~~y~~es~~constante:~~\frac{df}{dx}=\sin{y}-y\sin{x}
\end{align*}
$$

Integrando la última expresión:

$$
\begin{equation}
\begin{aligned}
f(x,y) &= \int (\sin{y}-y\sin{x}) dx + \color{green}{g(y)}\\
f(x,y) &= \int \sin{y}~dx - \int y\sin{x}~dx + \color{green}{g(y)}\\
f(x,y) &= x\sin{y} +y\cos{x}+ \color{green}{g(y)}
\end{aligned}
\label{ODE_EX_12}
\tag{28}
\end{equation}
$$

Ahora, derivando parcialmente con respecto a y:

$$
\begin{align*}
\left[\frac{\partial f(x,y)}{\partial y}\right]_{x} &= \frac{\partial}{\partial y}\left[x\sin{y} +y\cos{x}+ \color{green}{g(y)}\right]\\
\left[\frac{\partial f(x,y)}{\partial y}\right]_{x} &= \frac{\partial}{\partial y}(x\sin{y})+\frac{\partial}{\partial y}(y\cos{x})+\color{blue}{g'(y)}\\
\left[\frac{\partial f(x,y)}{\partial y}\right]_{x} &= x\cos{y}+\cos{x}+\color{blue}{g'(y)}
\end{align*}
$$

Recordando que $\left(\frac{\partial f}{\partial y}\right)_{x}= N(x,y)$, se tiene:

$$
\begin{align*}
\left[\frac{\partial f(x,y)}{\partial y}\right]_{x} &= x\cos{y}+\cos{x}+\color{blue}{g'(y)} = N(x,y)\\
\left[\frac{\partial f(x,y)}{\partial y}\right]_{x} &= x\cos{y}+\cos{x}+\color{blue}{g'(y)} = \cos{x}+x\cos{y}-y\\
\color{blue}{\frac{dg(y)}{dy}} &= -y\\
\color{green}{\int}\frac{dg(y)}{dy} \color{green}{dy} &= \color{green}{\int} -y~\color{green}{dy}\\
g(y) &=-\frac{y^2}{2}+c_{1}
\end{align*}
$$

Reemplazando este último resultado en la ecuación ([28](#mjx-eqn-ODE_EX_12)):
$$
\begin{align*}
f(x,y)&=c\\
 f(x,y) &= x\sin{y} +y\cos{x}-\frac{y^2}{2}+c_{1}=c\\
 f(x,y) &= x\sin{y} +y\cos{x}-\frac{y^2}{2}=c_{2}~~\Rightarrow~~\color{green}{c_{2}}= \color{green}{c-c_{1}}
\end{align*}
$$

La gráfica de la familia de soluciones se muestra a continuación:

```python
def f(x,y):
    return x*np.sin(y)+y*np.cos(x)
X,Y = np.meshgrid(np.linspace(-10,10,100),np.linspace(-10,10,100))
c=[-4,-2,0,2,4]
Z = f(X,Y)
fig = plt.figure(figsize=(15,5))
ax1 = fig.add_subplot(1,2,1)
cs = ax1.contour(X,Y,f(X,Y),c)
ax1.clabel(cs,inline=1,fontsize=8)
ax1.set_title("Fig.12: Gráfica de contorno de la familia de soluciones de la EDO 27", size = 14)
ax1.set_xlabel('x', size = 14)
ax1.set_ylabel('y',size = 14)

ax2 = fig.add_subplot(1,2,2,projection='3d')
ax2.plot_surface(X,Y,Z,cmap='viridis',rstride=2, cstride=2)
ax2.contour(X,Y,f(X,Y),c,offset=-20)
ax2.set_title("Fig.13: Gráfica 3D de la familia de soluciones de la EDO 27", size = 14)
ax2.set_zlim(-20,20)
ax2.set_xlabel('x', size = 14)
ax2.set_ylabel('y',size = 14)
ax2.set_zlabel('f(x,y)',size = 14)
plt.show()
```

Considere otro ejemplo:

$$
\begin{equation}
\begin{aligned}
\left(1+\ln{x}+\frac{y}{x}\right)dx-(1-\ln{x})dy=0
\end{aligned}
\label{ODE_EX_13}
\tag{29}
\end{equation}
$$

Para comprobar que el lado izquierdo de la ecuación ([29](#mjx-eqn-ODE_EX_13)) es un diferencial exacto, aplicamos el criterio dado por la ecuación ([21](#mjx-eqn-ODE_EX_5)):

$$
\begin{align*}
M(x,y) &= 1+\ln{x}+\frac{y}{x} \\
\frac{\partial M}{\partial y} &= \frac{\partial}{\partial y}(1)+\frac{\partial}{\partial y}(\ln{x})+\frac{\partial}{\partial y}\left(\frac{y}{x}\right)\\
\frac{\partial M}{\partial y} &= 0+0+\frac{1}{x}\\
\frac{\partial M}{\partial y} &= \frac{1}{x}
\end{align*}
$$

$$
\begin{align*}
N(x,y) &= -(1-\ln{x})=-1+\ln{x}\\
\frac{\partial N}{\partial x} &= \frac{\partial}{\partial x}(-1)+\frac{\partial}{\partial x}(\ln{x})\\
\frac{\partial N}{\partial x} &= 0+\frac{1}{x}\\
\frac{\partial N}{\partial x} &= \frac{1}{x}
\end{align*}
$$

De esto se concluye que $\frac{\partial M}{\partial y}=\frac{\partial N}{\partial x}$, cumpliéndose el criterio dado por la ecuación ([21](#mjx-eqn-ODE_EX_5)). Por tanto, el lado izquierdo de la ecuación ([29](#mjx-eqn-ODE_EX_13)) es un diferencial exacto, y entoces la ecuación ([29](#mjx-eqn-ODE_EX_13)) es exacta. 


$$
\begin{align*}
\left(\frac{\partial f}{\partial y}\right)_{x}=-1+\ln{x}~~\Rightarrow~~si~~x~~es~~constante:~~\frac{df}{dy}=-1+\ln{x}
\end{align*}
$$

Integrando:

$$
\begin{equation}
\begin{aligned}
f(x,y) &= \int (-1+\ln{x}) dy + \color{green}{h(x)}\\
f(x,y) &= -\int 1~dy + \int \ln{x}~dy + \color{green}{h(x)}\\
f(x,y) &= -y+y\ln{x}+\color{green}{h(x)}
\end{aligned}
\label{ODE_EX_14}
\tag{30}
\end{equation}
$$

Ahora, derivando parcialmente con respecto a x:

$$
\begin{align*}
\left[\frac{\partial f(x,y)}{\partial x}\right]_{y} &= \frac{\partial}{\partial x}\left[-y+y\ln{x}+\color{green}{h(x)}\right]\\
\left[\frac{\partial f(x,y)}{\partial x}\right]_{y} &= \frac{\partial}{\partial x}(-y)+\frac{\partial}{\partial y}(y\ln{x})+\color{blue}{h'(x)}\\
\left[\frac{\partial f(x,y)}{\partial x}\right]_{y} &= 0+\frac{y}{x}+\color{blue}{h'(x)}
\end{align*}
$$

Recordando que $\left(\frac{\partial f}{\partial x}\right)_{y}= M(x,y)$, entonces:

$$
\begin{align*}
\left[\frac{\partial f(x,y)}{\partial x}\right]_{y} &= \frac{y}{x}+\color{blue}{h'(x)} = M(x,y)\\
\left[\frac{\partial f(x,y)}{\partial x}\right]_{y} &= \frac{y}{x}+\color{blue}{h'(x)} = 1+\ln{x}+\frac{y}{x}\\
\color{blue}{\frac{dh(x)}{dx}} &= 1+\ln{x} \\
\color{green}{\int}\frac{dh(x)}{dx} \color{green}{dx} &= \color{green}{\int} 1~\color{green}{dx}+\color{green}{\int} \ln(x)~\color{green}{dx}\\
h(x) &=x+x\ln{x}-x+c_{1}\\
h(x) &= x\ln{x}+c_{1}
\end{align*}
$$

Reemplazando este último resultado en la ecuación ([30](#mjx-eqn-ODE_EX_14)):
$$
\begin{align*}
f(x,y) &=c\\
 f(x,y) &= -y+y\ln{x}+x\ln{x}+c_{1}=c\\
 f(x,y) &=-y+y\ln{x}+x\ln{x} =c_{2}~~\Rightarrow~~\color{green}{c_{2}}= \color{green}{c-c_{1}}
\end{align*}
$$

<h2><span style="color:blue" >2.4.6 Problema de valor inicial<span></h2>

Considere la siguiente ecuación diferencial:

$$
\begin{equation}
\begin{aligned}
(-y^{2}-y\sin{x})dx+\left(\frac{1}{1+y^{2}}+\cos{x}-2xy\right)=0
\end{aligned}
\label{ODE_EX_15}
\tag{31}
\end{equation}
$$

Sujeta a la condición inicial y(0)=1, es decir, el punto (0,1). En primer lugar, tiene que obtenerse la familia de soluciones de la EDO ([31](#mjx-eqn-ODE_EX_15)) aplicando el procedimiento visto anteriormente.

Es trivial demostrar que $\frac{\partial M}{\partial y}=\frac{\partial N}{\partial x}=-2y-\sin{x}$, satisfaciendo por tanto el criterio para que la la EDO ([31](#mjx-eqn-ODE_EX_15)) sea considerada como exacta.


$$
\begin{align*}
\left(\frac{\partial f}{\partial x}\right)_{y}=-y^{2}-y\sin{x}~~\Rightarrow~~si~~y~~es~~constante:~~\frac{df}{dx}=-y^{2}-y\sin{x}
\end{align*}
$$

Integrando la expresión anterior:

$$
\begin{equation}
\begin{aligned}
f(x,y) &= \int (-y^{2}-y\sin{x}) dx + \color{green}{g(y)}\\
f(x,y) &= -\int y^2~dx - \int y\sin{x}~dx + \color{green}{g(y)}\\
f(x,y) &= -xy^{2} +y\cos{x}+ \color{green}{g(y)}
\end{aligned}
\label{ODE_EX_16}
\tag{32}
\end{equation}
$$

Derivando $f(x,y)$ con respecto a y:

$$
\begin{align*}
\left[\frac{\partial f(x,y)}{\partial y}\right]_{x} &= \frac{\partial}{\partial y}\left[-xy^{2} +y\cos{x}+\color{green}{g(y)}\right]\\
\left[\frac{\partial f(x,y)}{\partial y}\right]_{x} &= \frac{\partial}{\partial y}(-xy^{2})+\frac{\partial}{\partial y}(y\cos{x})+\color{blue}{g'(y)}\\
\left[\frac{\partial f(x,y)}{\partial y}\right]_{x} &= -2xy+\cos{x}+\color{blue}{g'(y)}
\end{align*}
$$

Recordando que $\left(\frac{\partial f}{\partial y}\right)_{x}= N(x,y)$, se tiene:

$$
\begin{align*}
\left[\frac{\partial f(x,y)}{\partial y}\right]_{x} &= -2xy+\cos{x}+\color{blue}{g'(y)} = N(x,y)\\
\left[\frac{\partial f(x,y)}{\partial y}\right]_{x} &= -2xy+\cos{x}+\color{blue}{g'(y)} = \frac{1}{1+y^{2}}+\cos{x}-2xy\\
\color{blue}{\frac{dg(y)}{dy}} &= \frac{1}{1+y^{2}}\\
\color{green}{\int}\frac{dg(y)}{dy} \color{green}{dy} &= \color{green}{\int} \frac{1}{1+y^{2}}~\color{green}{dy}\\
g(y) &=\arctan(y)+c_{1}
\end{align*}
$$

Reemplazando este último resultado en la ecuación ([32](#mjx-eqn-ODE_EX_16)):
$$
\begin{align*}
f(x,y) &=c\\
 f(x,y) &= -xy^{2}+y\cos{x}+\arctan(y)+c_{1}=c\\
 f(x,y) &= -xy^{2}+y\cos{x}+\arctan(y) =c_{2}~~\Rightarrow~~\color{green}{c_{2}}= \color{green}{c-c_{1}}
\end{align*}
$$

Una vez obtenida la familia de soluciones de la EDO exacta ([31](#mjx-eqn-ODE_EX_15)), se reemplaza el punto $(0,1)$ y se obtiene el valor de la constante $c_{2}$:
$$
\begin{align*}
-xy^{2}+y\cos{x}+\arctan(y) &=c_{2}\\
-(0)(1^{2})+1[\cos(0)]+\arctan(1) &=c_{2}\\
0+1+\frac{\pi}{4} &= c_{2}\\
1+\frac{\pi}{4} &= c_{2}
\end{align*}
$$

Por tanto, la solución particular es:

$$
\begin{align*}
-xy^{2}+y\cos{x}+\arctan(y) = 1+\frac{\pi}{4}
\end{align*}
$$

En la Figura 16 se muestra la familia de soluciones de la EDO ([31](#mjx-eqn-ODE_EX_15)) y la solución particular que pasa por el punto $(0,1)$. En la Figura 17 se observa la superficie generada por la familia de soluciones y el punto que corresponde a las condiciones iniciales. 

```python
def f(x,y):
    return -x*y**2+y*np.cos(x)+np.arctan(y)
X,Y = np.meshgrid(np.linspace(-10,10,100),np.linspace(-5,5,100))
c= np.linspace(-10,10,20)
Z = f(X,Y)
fig = plt.figure(figsize=(15,5))
ax1 = fig.add_subplot(1,2,1)
cs = ax1.contour(X,Y,f(X,Y),c)
ax1.clabel(cs,inline=1,fontsize=8)
ax1.set_title("Fig.16: Gráfica de contorno de la familia de soluciones de la EDO 31", size = 14)
ax1.set_xlabel('x', size = 14)
ax1.set_ylabel('y',size = 14)
ax1.scatter(0,1, color='black')
ax1.text(0,1, '({}, {})'.format(0,1), ha='left',va='bottom',size=14)

ax2 = fig.add_subplot(1,2,2,projection='3d')
ax2.plot_surface(X,Y,Z,cmap='viridis',rstride=4, cstride=4)
ax2.contour(X,Y,f(X,Y),c,offset=-250)
ax2.set_title("Fig.17: Gráfica 3D de la familia de soluciones de la EDO 31", size = 14)
ax2.plot(0,1,1.785,color='red', marker='o')
ax2.text(0,1,1.785,'({}, {}, {})'.format(0,1,1.785), ha='left',va='bottom',size=14)
ax2.set_zlim(-250,250)
ax2.set_xlabel('x', size = 14)
ax2.set_ylabel('y',size = 14)
ax2.set_zlabel('f(x,y)',size = 14)
plt.show()
```

<h1><span style="color:purple">2.5 EDO de primer orden que se resuelven por sustitución<span></h1>

<h2><span style="color:blue" >2.5.1 Ecuaciones de grado homogéneo<span></h2>

<h3><span style="color:green" >a. Funciones homogéneas de grado  &#945;<span></h3>

Si una función $f$ tiene la propiedad:
$$
\begin{equation}
\begin{aligned}
f(tx,ty)=t^{\alpha}f(x,y)
\end{aligned}
\label{ODE_SUS_01}
\tag{33}
\end{equation}
$$

entonces se dice que es una función homogénea de grado $\alpha$. Por ejemplo, considere la función $f(x,y)=x^{2}-xy$. Para comprobar si esta es una función homogénea, sustituya $x\rightarrow tx$ y $y\rightarrow ty$, y obtenga una expresiíon similar a la ecuación ([33](#mjx-eqn-ODE_SUS_01)):

$$
\begin{align*}
\color{blue}{f(x,y)} &= \color{blue}{x^{2}-xy}\\
\color{green}{f(tx,ty)} &= (\color{green}{tx})^2-(\color{green}{tx})(\color{green}{ty})\\
\color{green}{f(tx,ty)} &= \color{green}{t^{2}}x^{2}-\color{green}{t^{2}}xy\\
\color{green}{f(tx,ty)} &= \color{green}{t^{2}}\color{blue}{(x^{2}-xy)} = \color{green}{t^{\alpha}}\color{blue}{f(x,y)}\\
\end{align*}
$$

De las operaciones realizadas se puede notar que la expresión $f(x,y)=x^{2}-xy$ es una función homogénea de grado 2, debido a que $\alpha=2$.

Por otra parte, considere la función $f(x,y)=x^{2}y-5$. Para comprobar si esta es homogénea se sigue el procedimiento detallado anteriormente:

$$
\begin{align*}
\color{blue}{f(x,y)} &= \color{blue}{x^{2}y-5}\\
\color{green}{f(tx,ty)} &= (\color{green}{tx})^2(\color{green}{ty})-5\\
\color{green}{f(tx,ty)} &= \color{green}{t^{3}}x^{2}y-5\neq\color{green}{t^{\alpha}}\color{blue}{f(x,y)}
\end{align*}
$$

Por tanto, se concluye que la función $f(x,y)=x^{2}y-5$ no es homogénea.

<h3><span style="color:green" >b. Ecuaciones diferenciales de grado homogeneo<span></h3>

Una ecuación diferencial ordinaria de primer orden expresada en la forma $M(x,y)dx+N(x,y)dy=0$ se dice que es <b>homogénea</b> si tanto $M(x,y)$ y $N(x,y)$ son funciones homogéneas del mismo grado. 

Sin este es el caso, se pueden plantear una de las siguientes sustituciones para resolver la EDO homogénea:
$$
\begin{equation}
\begin{aligned}
y &= ux\\
x &= vy
\end{aligned}
\label{ODE_SUS_02}
\tag{34}
\end{equation}
$$

<h3><span style="color:green" >c. Algoritmo de solución<span></h3>

<span style="color:#ff7f0e"><b>c.1 Comprobar criterio de homogeneidad</b></span>

Para esto, verifique que $M(x,y)$ y $N(x,y)$ sean funciones homogéneas del mismo grado:

$$
\begin{equation}
\begin{aligned}
M(tx,ty) &= t^{\alpha}M(x,y)\\
N(tx,ty) &= t^{\alpha}N(x,y)\\
\end{aligned}
\label{ODE_SUS_03}
\tag{35}
\end{equation}
$$

<span style="color:#ff7f0e"><b>c.2 Plantear una sustitución adecuada</b></span>

Escoja la sustitución que mejor permita resolver el problema de entre las planteadas en al ecuación ([34](#mjx-eqn-ODE_SUS_02)). Si toma $x=vy$ tenga en cuenta que $dx=vdy+ydv$, mientras que si escoge $y=ux$, entonces $dy=udx+xdu$.

Por ejemplo, asumiendo que se ha comprobado que $M(x,y)$ y $N(x,y)$ son funciones homogéneas del mismo grado, haciendo que $t=\frac{1}{x}$ y empleando la sustitución $y=ux$, se logra que las ecuaciones ([35](#mjx-eqn-ODE_SUS_03)) se transforman en:

$$
\begin{align*}
M(\color{blue}{t}x,\color{blue}{t}y) &= \color{blue}{t^{\alpha}}M(x,y)\\
M\left(\color{blue}{\frac{1}{x}}x,\color{blue}{\frac{1}{x}}y\right) &= \left(\color{blue}{\frac{1}{x}}\right)^{\alpha}M(x,y)\\
x^{\alpha}M\left(1,\frac{y}{x}\right) &= M(x,y)\\
x^{\alpha}M\left(1,u\right) &= M(x,y)
\end{align*}
$$

Adicionalmente:

$$
\begin{align*}
N(\color{green}{t}x,\color{green}{t}y) &= \color{green}{t^{\alpha}}N(x,y)\\
N\left(\color{green}{\frac{1}{x}}x,\color{green}{\frac{1}{x}}y\right) &= \left(\color{green}{\frac{1}{x}}\right)^{\alpha}N(x,y)\\
x^{\alpha}N\left(1,\frac{y}{x}\right) &= N(x,y)\\
x^{\alpha}N\left(1,u\right) &= N(x,y)
\end{align*}
$$

Por tanto, reemplazando los resultados anteriores en la EDO de primer orden que se quiere resolver:

$$
\begin{align*}
M(x,y)dx+N(x,y)dy &= 0\\
x^{\alpha}M\left(1,u\right)dx+x^{\alpha}N\left(1,u\right)dy &= 0\\
x^{\alpha}[M\left(1,u\right)dx+N\left(1,u\right)dy] &= 0\\
\color{blue}{M\left(1,u\right)dx+N\left(1,u\right)dy} &=\color{blue}{0},~~\color{green}{x^{\alpha}} = \color{green}{0}
\end{align*}
$$

Tomando la expresión en color azul y teniendo en cuenta que $dy=udx+xdu$:

$$
\begin{align*}
\color{blue}{M\left(1,u\right)dx+N\left(1,u\right)dy} &=\color{blue}{0}\\
M\left(1,u\right)dx+N\left(1,u\right)\color{green}{(udx+xdu)} &= 0\\
M\left(1,u\right)dx+\color{green}{u}N\left(1,u\right)\color{green}{dx}+\color{green}{x}N\left(1,u\right)\color{green}{du} &= 0\\
[M\left(1,u\right)+\color{green}{u}N\left(1,u\right)]dx &=-\color{green}{x}N\left(1,u\right)\color{green}{du}\\
\frac{dx}{x} &= -\frac{N\left(1,u\right)\color{green}{du}}{M\left(1,u\right)+\color{green}{u}N\left(1,u\right)}
\end{align*}
$$

Es decir, el resultado de la sustitución produce una EDO de primer orden que es posible resolver mediante <b>separación de variables</b>. Se llega a una expresión similar si se parte haciendo que $t=\frac{1}{y}$ y empleando la sustitución $x=vy$:

$$
\begin{align*}
M(\color{blue}{t}x,\color{blue}{t}y) &= \color{blue}{t^{\alpha}}M(x,y)\\
M\left(\color{blue}{\frac{1}{y}}x,\color{blue}{\frac{1}{y}}y\right) &= \left(\color{blue}{\frac{1}{y}}\right)^{\alpha}M(x,y)\\
y^{\alpha}M\left(\frac{x}{y},1\right) &= M(x,y)\\
y^{\alpha}M\left(v,1\right) &= M(x,y)
\end{align*}
$$

De manera similar:

$$
\begin{align*}
N(\color{green}{t}x,\color{green}{t}y) &= \color{green}{t^{\alpha}}N(x,y)\\
N\left(\color{green}{\frac{1}{y}}x,\color{green}{\frac{1}{y}}y\right) &= \left(\color{green}{\frac{1}{y}}\right)^{\alpha}N(x,y)\\
y^{\alpha}N\left(\frac{x}{y},1\right) &= N(x,y)\\
y^{\alpha}N\left(v,1\right) &= N(x,y)
\end{align*}
$$

Reemplazando estos resultados en la EDO de primer orden que se quiere resolver:

$$
\begin{align*}
M(x,y)dx+N(x,y)dy &= 0\\
y^{\alpha}M\left(v,1\right)dx+y^{\alpha}N\left(v,1\right)dy &= 0\\
y^{\alpha}[M\left(v,1\right)dx+N\left(v,1\right)dy] &= 0\\
\color{blue}{M\left(v,1\right)dx+N\left(v,1\right)dy} &=\color{blue}{0},~~\color{green}{y^{\alpha}} = \color{green}{0}
\end{align*}
$$

Tomando la expresión en color azul y teniendo en cuenta que $dx=vdy+ydv$:

$$
\begin{align*}
\color{blue}{M\left(v,1\right)dx+N\left(v,1\right)dy} &=\color{blue}{0}\\
M\left(v,1\right)\color{green}{(vdy+ydv)}+N\left(v,1\right)dy &= 0\\
\color{green}{v}M\left(v,1\right)\color{green}{dy}+\color{green}{y}M\left(v,1\right)\color{green}{dv}+N\left(v,1\right)dy &= 0\\
[\color{green}{v}M\left(v,1\right)+N\left(v,1\right)]dy &=-\color{green}{y}M\left(v,1\right)\color{green}{dv}\\
\frac{dy}{y} &= -\frac{M\left(v,1\right)\color{green}{dv}}{\color{green}{v}M\left(v,1\right)+N\left(v,1\right)}
\end{align*}
$$

Al igual que en el caso anterior, la sustitución produce una EDO de primer orden que es posible resolver mediante <b>separación de variables</b>.

<h3><span style="color:green" >d. Ejemplo<span></h3>

Considere la siguiente ecuación diferencial:

$$
\begin{equation}
\begin{aligned}
(y^{2}+yx)dx+x^{2}dy = 0
\end{aligned}
\label{ODE_SUS_04}
\tag{36}
\end{equation}
$$

El primer paso consiste en comprobar que $M(x,y)$ y $N(x,y)$ son funciones homogéneas del mismo grado. Reemplazando $x\rightarrow tx$ y $y\rightarrow ty$ se deberían obtener un par de ecuaciones similares a las expresadas en ([35](#mjx-eqn-ODE_SUS_03)):

$$
\begin{align*}
M(x,y) &= y^{2}+yx\\
M(tx,ty) &= (ty)^{2}+(ty)(tx)\\
M(tx,ty) &= t^{2}y^{2}+t^{2}yx\\
M(tx,ty) &= t^{2}(y^{2}+yx)=t^{\alpha}M(x,y)~~\Rightarrow \color{blue}{\alpha}=\color{blue}{2}
\end{align*}
$$

$$
\begin{align*}
N(x,y) &= x^{2}\\
N(tx,ty) &= (tx)^{2}\\
N(tx,ty) &= t^{2}x^{2}=t^{\alpha}N(x,y)~~\Rightarrow \color{blue}{\alpha}=\color{blue}{2}
\end{align*}
$$

Como se puede observar, $M(x,y)$ y $N(x,y)$ son funciones homogéneas del mismo grado $\alpha=2$. 

Luego, planteando la sustitución $y=ux$, y tomando en cuenta que $dy=udx+xdu$ se tiene:

$$
\begin{align*}
(y^{2}+yx)dx+x^{2}dy &= 0\\
[\color{green}{(ux)^{2}}+\color{green}{(ux)}x]dx+x^{2}\color{green}{(udx+xdu)} &=0\\
u^{2}x^{2}dx + ux^{2}dx + ux^{2}dx +x^{3}du &= 0\\
x^{2}dx{u^{2}+2u} &= -x^{3}du\\
\frac{x^{2}dx}{x^{3}} &= -\frac{du}{u^{2}+2u}\\
\color{blue}{\frac{dx}{x}} &= \color{blue}{-\frac{du}{u(u+2)}}\\
\end{align*}
$$

La última expresión en azul es una EDO de primer orden que se puede resolver por variables separables. Antes, es menester aplicar el método de fracciones parciales para resolver la expresión de lado derecho de la igualdad:

$$
\begin{align*}
\frac{1}{u(u+2)}= \frac{A}{u}+\frac{B}{u+2}
\end{align*}
$$

Multiplicando ambos miembros de la igualdad por $u(u+2)$ se obtiene:

$$
\begin{align*}
1= A(u+2)+Bu
\end{align*}
$$

La última expresión es una identidad, que puede verificarse para cualquier valor de $u$. Entonces:

$$
\begin{align*}
1 &= A(u+2)+Bu\\
u &=0~\Rightarrow~1=2A+0~\Rightarrow~A=\frac{1}{2}\\
u &=-2~\Rightarrow~1=0-2B~\Rightarrow~B=-\frac{1}{2}
\end{align*}
$$

Por tanto, las fracciones parciales obtenidas son:

$$
\begin{align*}
\frac{1}{u(u+2)}= \frac{1}{2}\frac{1}{u}-\frac{1}{2}\frac{1}{u+2}
\end{align*}
$$

Reemplazando estos resultados en la EDO obtenida por sustitución y realizando operaciones:

$$
\begin{align*}
\color{blue}{\frac{dx}{x}} &= \color{blue}{-\frac{du}{u(u+2)}}\\
\frac{dx}{x} &= -\frac{1}{2}\frac{du}{u}+\frac{1}{2}\frac{du}{u+2}\\
\int \frac{dx}{x} &= -\frac{1}{2}\int\frac{1}{u} + \frac{1}{2}\int\frac{du}{u+2}\\
\ln{|x|} &= -\frac{1}{2}\ln{|u|} +\frac{1}{2}\ln{|u+2|}+c\\
\end{align*}
$$

Multiplicando ambos miembros de la igualdad por 2:

$$
\begin{align*}
2\ln{|x|} &= -\ln{|u|} + \ln{|u+2|} +c_{1}~\Rightarrow~\color{blue}{c_{1}}=\color{blue}{2c}
\end{align*}
$$

Agrupando logaritmos y realizando operaciones:

$$
\begin{align*}
2\ln{|x|}+\ln{|u|}-\ln{|u+2|}  &= c_{1}\\
\ln{|x|^{2}}+\ln{\frac{|u|}{|u+2|}} &= c_{1}\\
\ln{\left(|x|^{2}\frac{|u|}{|u+2|}\right)} &= c_{1}\\
e^{\ln{\left(|x|^{2}\frac{|u|}{|u+2|}\right)}} &= e^{c_{1}}\\
|x|^{2}\frac{|u|}{|u+2|} &= c_{2}~\Rightarrow~\color{blue}{c_{2}}=\color{blue}{e^{c_{1}}}\\
x^{2}\frac{u}{u+2} &= c_{3}~\Rightarrow~\color{blue}{c_{3}}=\color{blue}{\pm{c_{2}}}\\
\end{align*}
$$

Finalmente, reemplazando $u = \frac{y}{x}$ en la última ecuación se tiene:

$$
\begin{align*}
x^{2}\frac{u}{u+2} &= c_{3}\\
x^{2}\frac{\frac{y}{x}}{\frac{y}{x}+2} &= c_{3}\\
x^{2}\frac{\frac{y}{x}}{\frac{y+2x}{x}} &= c_{3}\\
\frac{x^{2}y}{y+2x} &= c_{3}\\
\color{green}{x^{2}y} &= \color{green}{c_{3}(y+2x)}
\end{align*}
$$

Esta es la familia de soluciones de la EDO ([36](#mjx-eqn-ODE_SUS_04)).

<h2><span style="color:blue" >2.5.2 Ecuación de Bernoulli<span></h2>

<h3><span style="color:green" >a. Definición<span></h3>

Una ecuación diferencial de la forma:

$$
\begin{equation}
\begin{aligned}
\frac{dy}{dx}+P(x)y=f(x)y^{n}
\end{aligned}
\label{ODE_SUS_05}
\tag{37}
\end{equation}
$$

en donde $n$ es cualquier número real se denomina ecuación de Bernoulli. Para $n=0$ y $n=1$ la ecuación es lineal. Para el resto de valores de $n$, se recomienda la sustitución $u=y^{1-n}$ que reduce a la ecuación ([36](#mjx-eqn-ODE_SUS_04)) a una EDO lineal de primer orden.

<h3><span style="color:green" >b. Ejemplo<span></h3>

Como ejemplo, considere el siguiente problema de valor inicial:

$$
\begin{equation}
\begin{aligned}
x^{2}\frac{dy}{dx}-2xy=3y^{4}
\end{aligned}
\label{ODE_SUS_06}
\tag{38}
\end{equation}
$$

 sujeto a la condición inicial $y(1)=\frac{1}{2}$. Es importante notar que esta ecuación no está definida para $x=0$ puesto que en caso de hacerlo se eliminaría el operador diferencial.

<span style="color:#ff7f0e"><b>b.1 Colocar la EDO en su forma estándar</b></span>

Para esto, dividimos ambos miembros de la ecuación ([38](#mjx-eqn-ODE_SUS_06)) entre $x^{2}$:

$$
\begin{align*}
\frac{dy}{dx}-\frac{2xy}{x^{2}} &=\frac{3}{x^{2}}y^{4}\\
\frac{dy}{dx}-\frac{2}{x}y &= \frac{3}{x^{2}}y^{4}\\
\end{align*}
$$

<span style="color:#ff7f0e"><b>b.2 Comparar con la forma de la EDO de Bernoulli y obtener <i>n</i></b></span>

Para esto:

$$
\begin{align*}
\frac{dy}{dx}-\frac{2}{x}y &= \frac{3}{x^{2}}y^{4}\\
\frac{dy}{dx}+P(x)y &= f(x)y^{n}
\end{align*}
$$

Por comparación, $n=4$.

<span style="color:#ff7f0e"><b>b.3 Plantear la sustitución recomendada y resolver la ecuación</b></span>

La sustitución recomendada es $u=y^{1-n}$. Por tanto, para este caso $u=y^{1-4}=y^{-3}$. De aquí se puede ver que $y = u^{-1/3}$.

Ya que el cambio de variable es con respecto a la variable $u$, entonces es necesario determinar el valor de $\frac{du}{dx}$. Empleando la regla de la cadena: 

$$
\begin{align*}
\frac{dy}{dx}= \frac{dy}{du}\frac{du}{dx}
\end{align*}
$$

Ya que se determinó mediante la sustitución que $y =u^{-1/3}$, entonces $\frac{dy}{du}$ se podrá calcular mediante derivación:

$$
\begin{align*}
\frac{dy}{du} &= \frac{d}{du}(u^{-1/3})\\
\frac{dy}{du} &= -\frac{1}{3}u^{-4/3}
\end{align*}
$$

Por tanto:
$$
\begin{align*}
\frac{dy}{dx}= -\frac{1}{3}u^{-4/3}\frac{du}{dx}
\end{align*}
$$
    

Reemplazando los resultados obtenidos en la EDO $\frac{dy}{dx}-\frac{2}{x}y = \frac{3}{x^{2}}y^{4}$ se tiene:

$$
\begin{align*}
-\frac{1}{3}u^{-4/3}\frac{du}{dx}-\frac{2}{x}u^{-1/3}=\frac{3}{x^2}u^{-4/3}
\end{align*}
$$

Multiplicando ambos miembros de la última ecuación por $-3$ y dividiendo para $u^{-4/3}$, se llega a:

$$
\begin{align*}
\frac{du}{dx}+\frac{6}{x}u=-\frac{9}{x^2}
\end{align*}
$$

Claramente de puede observar que la última ecuación diferencial obtenida no está definida para $x=0$. También note que esta es una <b>EDO lineal</b>, por lo que para resolverla es necesario en primer lugar encontrar el factor de integración:

$$
\begin{align*}
P(x) &= \frac{6}{x}\\
w &= e^{\int P(x)dx}\\
w &= e^{\int \frac{6}{x} dx}\\
w &= e^{6\ln{|x|}}\\
w &= e^{\ln{|x|^{6}}}\\
w &= |x|^{6}
\end{align*}
$$

Por convención se puede tomar el valor positivo del valor absoluto en la última expresión, por lo cual el factor de integración es igual a $w=x^{6}$, y el intervalo de definición de la solución es $0<x<\infty$. Multiplicando el factor de integración en ambos lados de la ecuación diferencial lineal obtenida por sustitución y haciendo operaciones se llega a:

$$
\begin{align*}
x^{6}\frac{du}{dx}+x^{6}\frac{6}{x}u &= -x^{6}\frac{9}{x^2}\\
\frac{d}{dx}[x^{6}u] &= -9x^{4}\\
\int \frac{d}{dx}[x^{6}u] dx &= -\int 9x^{4} dx\\
x^{6}u &= -\frac{9}{5}x^{5}+c\\
u &= -\frac{9}{5} x^{-1} + cx^{-6}
\end{align*}
$$

Recordando que $u=y^{-3}$, se llega a la siguiente familia de soluciones uniparamétrica:

$$
\begin{align*}
y^{-3} &= -\frac{9}{5} x^{-1} + cx^{-6}
\end{align*}
$$

Sin embargo, el trabajo no termina ahí. Este es un problema de valor inicial por lo cual se requiere una solución particular. La condición inicial del problema es $(1,\frac{1}{2})$. Reemplazando esta para encontrar el valor de la constante se tiene:

$$
\begin{align*}
\left(\frac{1}{2}\right)^{-3} &= -\frac{9}{5} (1)^{-1} + c(1)^{-6}\\
\left(\frac{1}{2^{-3}}\right) &= -\frac{9}{5} + c\\
2^{3} + \frac{9}{5} &= c\\
8 + \frac{9}{5} &= c\\
\frac{49}{5} &= c
\end{align*}
$$

Por tanto, la solución particular buscada es:

$$
\begin{align*}
y^{-3} &= -\frac{9}{5} x^{-1} + \frac{49}{5}x^{-6}
\end{align*}
$$
